### Cartoon Box Company - Price Calculation:



#### Program Brief:

A manufacturing owner wants to calculate the prices of various Cartoon Boxes based on the inputs and conditions as specified by the owner. There are various inputs to be taken and based on the specifications on the desired cartoon box, the calculation and formulas for each type vary. The end result should display the size of the box, various calculation details with its specifications for manufacturing process based on the users inputs. 

The code should work in such a way that once the size of the Cartoon box is entered by the user with all the other box specifications required in the manufacturing, the program should display the details for the manufacturing to take place with all other pricing details and calculations.

Inputs from the Owner:

- Cartoon Box size: (Length, Breadth, Height) in Inches
- Number of Ply used: It can vary from (3,5,7,9,12,...)
- Paper Details: As mentioned by the owner (various options).
- Calculation Formulas: As mentioned by the owner (various options).




```python
# Python code:
# Asking the user for the inputs for the desired calculation:
print('Cartoon Box Company - Price Calculation')
print()
option = int(input('Choose the task you want to do from the below options: \n\n1. Calculating the Price of the Box \n\n2. Calculating the Paste-Chemical required \n\nEnter the option:'))
print()
# Asking the user for the option:
if 1 == option:
    print('Enter the box dimensions in Inches:')
    l = float(input('Please enter the Length: '))
    b = float(input('Please enter the Breath: '))
    h = float(input('Please enter the Height: '))
    rs = b+h+1
    cs = l+b+2
    print()
    print('Reel Size: ',rs)
    print('Cutting size: ',cs)
    print()
    print('Enter the paper quality used for rate claculation:')
    ply = int(input('Enter the no.of ply: '))
    gsm_base = int(input('Enter the Base GSM: '))
    gsm_flute = int(input('Enter the Flute GSM: '))
    bf = int(input('Enter the bf: '))
    rate = float(input('Enter the rate: ')) 
    flute_height = int(input("Enter the Flute Height category:"))
    fh = gsm_flute/flute_height
    
    
    # for 3-ply,120-gsm_base,120-gsm_flute,18-bf,rate-47.50:
    if ply == 3 and gsm_base == 120 and gsm_flute == 120 and bf == 18:
        weight = (((rs*cs*2)*fh)/1550)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
       
        
    # for 5-ply,120-gsm_base,120-gsm_flute,18-bf,rate-46.50:
    elif ply == 5 and gsm_base == 120 and gsm_flute == 120 and bf == 18: 
        weight = (((rs*cs*2)*720)/1550)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
        
    # for 7-ply,120-gsm_base,120-gsm_flute,18-bf,rate-46.50:
    elif ply == 7 and gsm_base == 120 and gsm_flute == 120 and bf == 18: 
        weight = (((rs*cs*2)*720)/1020)/1000
        price = weight*46.50
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
        
    # for 3-ply,150-gsm_base,150-gsm_flute,24-bf,rate-50:  
    elif ply == 3 and gsm_base == 150 and gsm_flute == 150 and bf == 24:
        weight = (((rs*cs*2)*420)/1550)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
        
    # for 5-ply,150-gsm_base,150-gsm_flute,24-bf,rate-50:     
    elif ply == 5 and gsm_base == 150 and gsm_flute == 150 and bf == 24: 
        weight = (((rs*cs*2)*720)/1550)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
        
    # for 7-ply,150-gsm_base,150-gsm_flute,24-bf,rate-50:    
    elif ply == 7 and gsm_base == 150 and gsm_flute == 150 and bf == 24:
        weight = (((rs*cs*2)*720)/1020)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
        
    # for 7-ply,150-gsm_base,150-gsm_flute,24-bf,rate-50:    
    elif ply == 7 and gsm_base == 150 and gsm_flute == 150 and bf == 24:
        weight = (((rs*cs*2)*720)/1020)/1000
        price = weight*rate
        print()
        print('Weight: ', weight)
        print('Price: ', price)
        print()
        no_of_boxes = int(input("Enter the No.of Boxes required: "))
        total_kgs = weight*no_of_boxes
        print("Total KGS of Paper required: ", total_kgs)
elif 2 == option:
    print('Enter the box dimensions in Inches:')
    l = float(input('Please enter the Length: '))
    b = float(input('Please enter the Breath: '))
    h = float(input('Please enter the Height: '))
    rs = b+h+1
    cs = l+b+2
    print()
    print('Reel Size: ',rs)
    print('Cutting size: ',cs)
    print()
    print('For Paste quality grade claculation:')
    paste_grade = int(input('Enter the grade: '))
    number_of_boxes = int(input('Enter the number of boxes required: '))
    paste_quantity = (paste_grade*(rs+cs)*number_of_boxes)/100000
    print("The amount of paste required is: ",paste_quantity, " kgs")
    
    # To get the correct input from the user:
else:
    option = int(input('Please enter the correct option: '))
    
```
